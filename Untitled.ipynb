```python
import requests
from git import Repo

def create_github_branch(repo_owner, repo_name, base_branch, new_branch):
    github_token = "YOUR_GITHUB_TOKEN"

    
    repo_url = f"https://github.com/{repo_owner}/{repo_name}.git"
    local_repo_path = f"{repo_name}_clone"
    Repo.clone_from(repo_url, local_repo_path)

    
    repo = Repo(local_repo_path)
    base_ref = repo.create_head(base_branch)
    new_ref = base_ref.checkout(branch=new_branch)

   
    source_folder = "path/to/source/folder"
    destination_folder = "path/to/destination/folder"
    repo.git.mv(source_folder, destination_folder)

    
    repo.index.commit(f"Move {source_folder} to {destination_folder}")
    repo.remotes.origin.push(new_branch)

    
    create_pull_request(repo_owner, repo_name, base_branch, new_branch)

def create_pull_request(repo_owner, repo_name, base_branch, new_branch):
    
    github_token = "YOUR_GITHUB_TOKEN"

    
    pr_url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/pulls"

    
    pr_data = {
        "title": f"Move folder and create PR",
        "head": new_branch,
        "base": base_branch,
        "body": "Please review the changes.",
    }

    headers = {
        "Authorization": f"Bearer {github_token}",
        "Accept": "application/vnd.github.v3+json",
    }

    
    response = requests.post(pr_url, json=pr_data, headers=headers)

    if response.status_code == 201:
        print("Pull request created successfully!")
    else:
        print(f"Failed to create pull request. Status code: {response.status_code}")
        print(response.text)

if __name__ == "__main__":
    repo_owner = "your_username"
    repo_name = "your_repo"
    base_branch = "main"
    new_branch = "feature/move-folder"

    create_github_branch(repo_owner, repo_name, base_branch, new_branch)
```
